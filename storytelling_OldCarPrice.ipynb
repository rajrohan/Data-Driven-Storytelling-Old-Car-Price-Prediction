```python
import numpy as np
import pandas as pd
```

```python
oldCar = "C:\\Master\\data\\github\\old-car\\data-prep\\FullDataset.csv"
```

```python
cardf = pd.read_csv(oldCar)
cardf.head()
```

```python
cardf = cardf.drop(['Name', 'Latitude (generated)','Longitude (generated)'], axis=1)
cardf.head()
```

```python
cardf = cardf[['Price','Year','Kilometers_Driven','Mileage','Engine','Power','Seats','Fuel_Type','Transmission','Owner_Type','Location','Brand']]
cardf.head()
```

```python
X = cardf[['Price','Year','Kilometers_Driven','Mileage','Engine','Power','Seats','Fuel_Type','Transmission','Owner_Type','Location','Brand']]
Y = cardf.Price
X = pd.get_dummies(data=X)
X.head()
```

```python
X.to_csv("C:\\Master\\data\\github\\old-car\\modelinput.csv")
```

```python
Y.head()
```

```python
X = X.drop(['Price'], axis=1)
X.head()
```

```python
X = X.drop(['Seats'], axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .30, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regr_1 = DecisionTreeRegressor(max_depth=5)
regr_2 = DecisionTreeRegressor(max_depth=7)
regr_1.fit(X_train,Y_train)
regr_2.fit(X_train,Y_train)
```

```python
y_pred = regr_1.predict(X_test)
y_pred2 = regr_2.predict(X_test)
```

```python
print("Accuracy: Depth 5",regr_1.score(X_test,Y_test))
print("Accuracy: Depth 7",regr_2.score(X_test,Y_test))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
lr_train_score=lr.score(X_train,Y_train)
lr_test_score=lr.score(X_test,Y_test)
print ("LR training score:", lr_train_score )
print ("LR test score: ", lr_test_score)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score

gbr = GradientBoostingRegressor(loss ='ls', max_depth=6)
gbr.fit (X_train, Y_train)
predicted = gbr.predict(X_test)
residual = Y_test - predicted
print("Accuracy:",gbr.score(X_test,Y_test))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(30,30))

ax2 = plt.subplot(212)
plt.scatter(predicted, residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.xlabel('Predicted',fontsize=25)
plt.ylabel('Residual',fontsize=25)
plt.axhline(y=0)
plt.title('Residual vs. Predicted',fontsize=35)

plt.show()
```
